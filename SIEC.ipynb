Finding Q1,Q3,IQR & whiskers for outlier analysis

```python
Q1 = df["Usage_kWh"].quantile(0.25)
Q3 = df["Usage_kWh"].quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
print(f"First quartile : {Q1}\nThird quartile : {Q3}\nInner quartile range(IQR) : {IQR}\nWhiskers : [{lower_limit},{upper_limit}]")
```

Dealing with catagorical data

Removing "WeekStatus_Weekday","WeekStatus_Weekend","Load_Type_Light_Load","Load_Type_Maximum_Load","Load_Type_Medium_Load","Lagging_Current_Power_Factor","Leading_Current_Power_Factor", because the data is already represented in other columns

```python
df = pd.get_dummies(df)
df = df.drop(
    [
        "WeekStatus_Weekday",
        "WeekStatus_Weekend",
        "Load_Type_Light_Load",
        "Load_Type_Maximum_Load",
        "Load_Type_Medium_Load",
        "Lagging_Current_Power_Factor",
        "Leading_Current_Power_Factor",
    ],
    axis=1,
)
df
```

<center><b><p style="font-size: 24px;">Building our models</p></b></center>


Training a Linear regression model and checking it's measures

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
X = df.drop(["Usage_kWh"], axis=1)
y = df.Usage_kWh
linear_regression_model = LinearRegression()
ten_fold = KFold(n_splits=10)
linear_models = []
linear_model_scores = []
linear_model_mae_scores = []
linear_model_mse_scores = []
linear_model_rmse_scores = []
for train_index, test_index in ten_fold.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    model = linear_regression_model.fit(X_train, y_train)
    linear_models.append(model)

    y_pred = model.predict(X_test)

    linear_model_scores.append(model.score(X_test, y_test))
    linear_model_mae_scores.append(mean_absolute_error(y_test, y_pred))
    linear_model_mse_scores.append(mean_squared_error(y_test, y_pred))
    linear_model_rmse_scores.append(np.sqrt(mean_squared_error(y_test, y_pred)))

best_model = np.argmax(linear_model_scores)

linear_model_score = round(linear_model_scores[best_model] * 100, 2)
linear_model_mae_score = round(linear_model_mae_scores[best_model], 2)
linear_model_mse_score = round(linear_model_mse_scores[best_model], 2)
linear_model_rmse_score = round(linear_model_rmse_scores[best_model], 2)

print(round(linear_model_scores[best_model] * 100, 2))
print(round(linear_model_mae_scores[best_model], 2))
print(round(linear_model_mse_scores[best_model], 2))
print(round(linear_model_rmse_scores[best_model], 2))
linear_regression_model = linear_models[best_model]
```

<center><b><p style="font-size: 24px;">Comparing our models</p></b></center>


Creating a dictionary with model measures

```python
metrics = {
    "Model": ["Linear", "Ridge", "Lasso", "LassoLars"],
    "R^2": [
        linear_model_score,
        ridge_model_score,
        lasso_model_score,
        lassolars_model_score,
    ],
    "MAE": [
        linear_model_mae_score,
        ridge_model_mae_score,
        lasso_model_mae_score,
        lassolars_model_mae_score,
    ],
    "MSE": [
        linear_model_mse_score,
        ridge_model_mse_score,
        lasso_model_mse_score,
        lassolars_model_mse_score,
    ],
    "RMSE": [
        linear_model_rmse_score,
        ridge_model_rmse_score,
        lasso_model_rmse_score,
        lassolars_model_rmse_score,
    ],
}
```

Using our dictionary to create a table and compare models

```python
mdf = pd.DataFrame(metrics)
mdf = mdf.set_index("Model")
mdf.T.style.set_table_styles(
    [{"selector": "th", "props": [("text-align", "center")]}]
).set_properties(**{"text-align": "center"}).format("{:.2f}")
```

We can see in the table above that Linear regression is the best model to use in this case
